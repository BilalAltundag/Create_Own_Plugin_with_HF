```python
import requests
import os

def download_image(url, save_directory):
    # Send a GET request to the URL
    response = requests.get(url)

    # Extract the filename from the URL
    filename = url.split("/")[-1]

    # Create the complete path to save the image
    save_path = os.path.join(save_directory, filename)

    # Save the image to the specified directory
    with open(save_path, "wb") as f:
        f.write(response.content)

    print(f"Image downloaded and saved as {filename} in {save_directory}")
```

## Let's Create Some Useful Tool

### **Get Customer Information Agent**

```python
# Assuming you have the DataFrame 'df' containing the customer information

def get_customer_info(name, last_name=None):
    if last_name is None:
        # Filter by name only
        filtered_df = df[df['Name'] == name]
    else:
        # Filter by name and last name
        filtered_df = df[(df['Name'] == name) & (df['Last Name'] == last_name)]

    if len(filtered_df) == 0:
        return "No customer found."
    else:
        return filtered_df.to_dict('records')
```

```python
import pandas as pd

data = [
    {"Name": "John", "Last Name": "Smith", "Working Status": "Employed", "Rating Score": 4},
    {"Name": "Emma", "Last Name": "Johnson", "Working Status": "Employed", "Rating Score": 3},
    {"Name": "Alex", "Last Name": "Brown", "Working Status": "Unemployed", "Rating Score": 2},
    {"Name": "Olivia", "Last Name": "Davis", "Working Status": "Employed", "Rating Score": 5},
    {"Name": "Olivia", "Last Name": "Brown", "Working Status": "Employed", "Rating Score": 6},
    {"Name": "Liam", "Last Name": "Wilson", "Working Status": "Employed", "Rating Score": 1}
]

df = pd.DataFrame(data)

print(df)
```

```python
from transformers import Tool
from huggingface_hub import list_models
from transformers.tools import HfAgent

class CostumerInfoTool(Tool):
    name = "costumer_info2"
    description = ("Returns Zoi customer information. It takes 'cust_name' and 'cust_last_name', and returns Zoi customer information.")

    inputs = ["text", "text"]
    outputs = ["text"]

    def __call__(self, cust_name: str, cust_last_name: str):
        for i in range(len(df)):
            if (df["Name"][i] in cust_name) or (df["Name"][i] in cust_name and df["Last Name"][i] in cust_last_name):
                return str(get_customer_info(df["Name"][i], df["Last Name"][i]))
        return "Customer not found"

tool = CostumerInfoTool()

agent = HfAgent(
    "https://api-inference.huggingface.co/models/bigcode/starcoder",
    token=token,
    additional_tools=[tool]
)

agent.run("get Olivia information from Zoi")
```

For test, you need to add your image in **images** file

```python
class FastSamTool(Tool):
    name = "FastSamTool"
    description = ("Returns image segmentation of image. It takes 'image' and 'text_prompt', and returns segmented image.")

    inputs = ["text", "text"]
    outputs = ["image"]

    def __call__(self, image: str, text_prompt: str):
        command = f"python FastSAM/Inference.py --model_path FastSAM.pt --img_path ./images/{image} --text_prompt '{text_prompt}' --imgsz 1024"
        # Run the command
        !{command}

        # Read the segmented image
        image = Image.open(f"./output/{image}")

        return image

tool = FastSamTool()

agent = HfAgent(
    "https://api-inference.huggingface.co/models/bigcode/starcoder",
    token=token,
    additional_tools=[tool]
)

agent.run("Do Fast Sam image segmentation", image="dog.jpg", text_prompt="A dog")
```
